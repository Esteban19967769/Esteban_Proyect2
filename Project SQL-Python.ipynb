```python
#Retrieve Metadata for the Database Server
server = ibm_db.server_info(conn)

print ("DBMS_NAME: ", server.DBMS_NAME)
print ("DBMS_VER:  ", server.DBMS_VER)
print ("DB_NAME:   ", server.DB_NAME)
```

```python
#Retrieve Metadata for the Database Client / Driver
client = ibm_db.client_info(conn)

print ("DRIVER_NAME:          ", client.DRIVER_NAME) 
print ("DRIVER_VER:           ", client.DRIVER_VER)
print ("DATA_SOURCE_NAME:     ", client.DATA_SOURCE_NAME)
print ("DRIVER_ODBC_VER:      ", client.DRIVER_ODBC_VER)
print ("ODBC_VER:             ", client.ODBC_VER)
print ("ODBC_SQL_CONFORMANCE: ", client.ODBC_SQL_CONFORMANCE)
print ("APPL_CODEPAGE:        ", client.APPL_CODEPAGE)
print ("CONN_CODEPAGE:        ", client.CONN_CODEPAGE)
```

### Problem 7: List 5 community areas with highest % of households below poverty line.


```python
%sql select community_area_name, percent_households_below_poverty from CENSUS_DATA order by percent_households_below_poverty DESC limit 5;
#selecciono las varibales, las ordeno por porcentaje de manera decreciente, el top 5
```

### Problem 8: Which community area(number) is most crime prone?


```python
%sql select community_area_number, count(*) as "frecuencias" from CRIME_DATA group by community_area_number order by "frecuencias" DESC limit 1; 
#selecciono la variable, le creo una nueva variable llamada frecuencias que son las muestras de cada dato, desde la base ordenada por la variable y ordenada de manera descendente por "frecuencias"
```
